Import numpy for numerical computations and matplotlib/pyplot for plotting

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', size=16)
```

Define functions

```python
def f1(x):
    return x**7 - 7*(x**6) + 21*(x**5) - 35*(x**4) + 35*(x**3) -21*(x**2) + 7*x - 1
```

Generate plot of function

```python
x = np.linspace(0.98 ,1.02,100)
y1 = f1(x)
fig, ax = plt.subplots()
ax.axis([0.98,1.02,-2e-14,2e-14])
ax.plot(x,y1,label='$x^7-7x^6+21x^5-35x^4+35x^3-21x^2+7x-1$')
ax.plot(x,np.linspace(0,0,100),label='$0$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

The above is quite worrisome, it doesn't look anything like a 7th-degree polynomial in $x$. We can see that the absolute errors are arguably small (see how the vertical axis is scaled by $10^{-14}$), but this would still be a significant issue if we were using this as the denominator of a fraction for instance.

We can get around some of these issues by rewriting the function in a way that would be equivalent (in $\mathbb{R}$), but will produce more accurate results in floating point numbers

```python
def f2(x):
    return (x-1)**7
```

```python
x = np.linspace(0.98 ,1.02,100)
y2 = f2(x)
fig, ax = plt.subplots()
ax.axis([0.98,1.02,-2e-14,2e-14])
ax.plot(x,np.linspace(0,0,100),'orange',label='$0$')
ax.plot(x,y1,'b', label='$x^7-7x^6+21x^5-35x^4+35x^3-21x^2+7x-1$')
ax.plot(x,y2,'r', label='$(x-1)^7$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```
